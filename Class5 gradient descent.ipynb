```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 

x = np.linspace(-1.,6.,141)
# 定义一个二次函数，作为损失函数
y = (x-2.5)**2-1
```

```python
plt.plot(x,y)
```

```python
epsilon = 1e-8 # 0
eta = 0.1 #步长（学习率）

def J(theta):
    return (theta-2.5)**2-1
def dJ(theta):
    return 2*(theta-2.5)
theta = 0
while True:
    last_theta = theta #上一步的theta值
    gradient = dJ(theta) #导数
    theta = theta- eta*gradient
    if(abs(J(theta)-J(last_theta))<epsilon):break

print(theta,J(theta)) # 对应图中的最低点
```

```python
# 可视化过程
theta= 0.0
eta = 0.1
history = [theta]
while True:
    last_theta = theta #上一步的theta值
    gradient = dJ(theta) #导数
    theta = theta- eta*gradient
    history.append(theta) #存储theta
    if(abs(J(theta)-J(last_theta))<epsilon):break

y_history = [J(i) for i in history ]

import seaborn as sns
sns.scatterplot(history,y_history,color = 'red',markers='x')
sns.lineplot(x,y)
```

```python
# 封装成函数
def Mygradient_decent(theta,eta,epsilon): #初始值、步长、最小值
    history = [theta]
    while True :
        last_theta = theta
        gradient = dJ(theta)
        theta = theta - eta * gradient
        history.append(theta)
        if(abs(J(theta)-J(last_theta))<epsilon):break
    y_history = [J(i) for i in history]
    return history,y_history

def Mypolt(history,y_history,eta):
    plt.figure(figsize=(5,5))
    plt.title('eta = '+"%.1f" % eta)
    sns.scatterplot(history,y_history,color = 'red',markers='x')
    sns.lineplot(x,y)
```

```python
# 查看不同eta 运行情况
epsilon = 1e-8
for i in range(5):
    eta = 0.1 + 0.1*i
    history,y_history = Mygradient_decent(0,eta,epsilon)
    Mypolt(history,y_history,eta)
```
